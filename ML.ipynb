<h1>Spilting of Data</h1>
<ol>
<li> X = Independent variable </li>
<li> Y = Dependent variable </li>
</ol>

```python
x = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']]
y = df[['Outcome']]

print(x)
print(y)
```

<h1>Training & Testing</h1>

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x, y, test_size= 0.2,random_state =42 )

print(X_train,Y_train.shape)
print(X_test,Y_test.shape)
```

<h1>Algorithms</h1>

<h1><li>Linear Regression</li></h1>

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)

pred = lr.predict(X_test)

#Score
print(lr.score(X_test,Y_test))

# Accuary score methods
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,accuracy_score,confusion_matrix

# MSE, MAE, r2score

print("MSE",mean_squared_error(Y_test,pred))
print("MAE",mean_absolute_error(Y_test,pred))
print("R2_score",r2_score(Y_test,pred))
```

<h1>Logistic Regression</h1>

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,Y_train)
LRpred = LR.predict(X_test)

print(LR.score(X_test,Y_test))
#Score ----> from sklearn.metrices import r2_score,mean_squared_error,mean_absolute_error
print("MSE",mean_squared_error(Y_test,LRpred))
print("MAE",mean_absolute_error(Y_test,LRpred))
print("R2_score",r2_score(Y_test,LRpred))

acc = accuracy_score(Y_test,LRpred)
print(acc)
cm = confusion_matrix(Y_test,LRpred)
print(cm)
```

<h1>Decision Tree</h1>

```python
from sklearn.tree import DecisionTreeClassifier
DTC =  DecisionTreeClassifier()
DTC.fit(X_train,Y_train)
DTC_pred = DTC.predict(X_test)

print(DTC.score(X_test,Y_test))
#Score ----> from sklearn.metrices import r2_score,mean_squared_error,mean_absolute_error
print("MSE",mean_squared_error(Y_test,DTC_pred))
print("MAE",mean_absolute_error(Y_test,DTC_pred))
print("R2_score",r2_score(Y_test,DTC_pred))

acc = accuracy_score(Y_test,DTC_pred)
print(acc)
cm = confusion_matrix(Y_test,DTC_pred)
print(cm)
```

<h1>Random Forest Classifier</h1>

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(X_train,Y_train)
RFCpred = RFC.predict(X_test)
print(RFC.score(X_test,Y_test))
#Score ----> from sklearn.metrices import r2_score,mean_squared_error,mean_absolute_error
print("MSE",mean_squared_error(Y_test,RFCpred))
print("MAE",mean_absolute_error(Y_test,RFCpred))
print("R2_score",r2_score(Y_test,RFCpred))

acc = accuracy_score(Y_test,RFCpred)
print(acc)
cm = confusion_matrix(Y_test,RFCpred)
print(cm)
```

<h1>SVM</h1>

```python
from sklearn.svm import SVC
SVM = SVC()
SVM.fit(X_train,Y_train)
SVM_Pred = SVM.predict(X_test)

print(SVM.score(X_test,Y_test))
#Score ----> from sklearn.metrices import r2_score,mean_squared_error,mean_absolute_error
print("MSE",mean_squared_error(Y_test,SVM_Pred))
print("MAE",mean_absolute_error(Y_test,SVM_Pred))
print("R2_score",r2_score(Y_test,SVM_Pred))

acc = accuracy_score(Y_test,SVM_Pred)
print(acc)
cm = confusion_matrix(Y_test,SVM_Pred)
print(cm)
```
